```python
# ============================================================
# ✅ BERT for Self-Fixed SATD Prediction (10-Fold + LOPO + EarlyStopping + )
# ============================================================
!pip install -q transformers==4.42.3 pandas scikit-learn tqdm

import json
import torch
import torch.nn as nn
import numpy as np
import pandas as pd
from tqdm import tqdm
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import (
    accuracy_score, f1_score, precision_score, recall_score, roc_auc_score
)
from torch.utils.data import Dataset, DataLoader
from transformers import (
    BertTokenizer, BertModel, AdamW, get_linear_schedule_with_warmup
)

# ============================================================
# 1. （）
# ============================================================
RAW_DATA = 'raw_data_final_40501.json'
OUT_CV   = 'bert_pytorch_10fold_earlystop.txt'
OUT_LOPO = 'bert_pytorch_lopo_earlystop.txt'
MODEL_NAME = './autodl-fs/bert-base-uncased'  # ✅ BERT

# 
MAX_LEN = 128
BATCH_SIZE = 32
EPOCHS = 50
LR = 2e-5
EMBED_DIM = 300
PATIENCE = 3
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# ============================================================
# 2. 
# ============================================================
print("📦 Loading dataset ...")
with open(RAW_DATA, 'r', encoding='utf-8') as f:
    records = json.load(f)
df = pd.DataFrame(records)
texts    = df['f_comment'].astype(str).tolist()
labels   = df['is_self_fixed'].astype(int).values
projects = df['project_name'].astype(str).tolist()
print(f"✅ Data loaded: {len(df)} samples from {df['project_name'].nunique()} projects.")

# ============================================================
# 3. Tokenizer & Dataset （）
# ============================================================
print("🧠 Loading local tokenizer ...")
tokenizer = BertTokenizer.from_pretrained(MODEL_NAME, local_files_only=True)

class SATDDataset(Dataset):
    def __init__(self, texts, labels):
        self.texts = texts
        self.labels = labels

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        text = str(self.texts[idx])
        label = self.labels[idx]
        encoded = tokenizer.encode_plus(
            text,
            add_special_tokens=True,
            max_length=MAX_LEN,
            truncation=True,
            padding='max_length',
            return_tensors='pt'
        )
        return {
            'input_ids': encoded['input_ids'].squeeze(),
            'attention_mask': encoded['attention_mask'].squeeze(),
            'label': torch.tensor(label, dtype=torch.long)
        }

# ============================================================
# 4. （）
# ============================================================
class BertForSATD(nn.Module):
    def __init__(self, num_labels=2, embed_dim=300):
        super(BertForSATD, self).__init__()
        self.bert = BertModel.from_pretrained(MODEL_NAME, local_files_only=True)
        self.down_proj = nn.Linear(768, embed_dim)
        self.dropout = nn.Dropout(0.3)
        self.classifier = nn.Linear(embed_dim, num_labels)

    def forward(self, input_ids, attention_mask, labels=None):
        outputs = self.bert(input_ids=input_ids, attention_mask=attention_mask)
        cls_output = outputs.last_hidden_state[:, 0, :]  # [CLS] 
        reduced = self.down_proj(cls_output)
        dropped = self.dropout(reduced)
        logits = self.classifier(dropped)
        if labels is not None:
            loss_fct = nn.CrossEntropyLoss()
            loss = loss_fct(logits, labels)
            return loss, logits
        else:
            return logits

# ============================================================
# 5. （）
# ============================================================
def evaluate_model(model, dataloader):
    model.eval()
    preds, probs, trues = [], [], []
    val_loss = 0
    loss_fn = nn.CrossEntropyLoss()

    with torch.no_grad():
        for batch in dataloader:
            input_ids = batch['input_ids'].to(DEVICE)
            attention_mask = batch['attention_mask'].to(DEVICE)
            labels = batch['label'].to(DEVICE)

            logits = model(input_ids, attention_mask)
            loss = loss_fn(logits, labels)
            val_loss += loss.item()

            prob = torch.softmax(logits, dim=1)[:, 1].cpu().numpy()
            preds_batch = torch.argmax(logits, dim=1).cpu().numpy()

            preds.extend(preds_batch)
            probs.extend(prob)
            trues.extend(labels.cpu().numpy())

    val_loss /= len(dataloader)
    acc = accuracy_score(trues, preds)
    f1 = f1_score(trues, preds)
    prec = precision_score(trues, preds)
    rec = recall_score(trues, preds)
    try:
        auc = roc_auc_score(trues, probs)
    except:
        auc = float('nan')

    return val_loss, acc, prec, rec, f1, auc


def train_one_fold(train_idx, val_idx, texts, labels, fold_id=None, tag="CV"):
    train_dataset = SATDDataset([texts[i] for i in train_idx], labels[train_idx])
    val_dataset   = SATDDataset([texts[i] for i in val_idx], labels[val_idx])

    train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
    val_loader   = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False)

    model = BertForSATD(num_labels=2, embed_dim=EMBED_DIM).to(DEVICE)
    optimizer = AdamW(model.parameters(), lr=LR)
    total_steps = len(train_loader) * EPOCHS
    scheduler = get_linear_schedule_with_warmup(optimizer, 0, total_steps)

    best_metrics = {'val_loss': float('inf'), 'acc': 0, 'prec': 0, 'rec': 0, 'f1': 0, 'auc': 0}
    patience_counter = 0
    best_model_state = None

    for epoch in range(EPOCHS):
        model.train()
        total_loss = 0
        for batch in tqdm(train_loader, desc=f"{tag} {fold_id} | Epoch {epoch+1}/{EPOCHS}"):
            optimizer.zero_grad()
            input_ids = batch['input_ids'].to(DEVICE)
            attention_mask = batch['attention_mask'].to(DEVICE)
            labels = batch['label'].to(DEVICE)

            loss, logits = model(input_ids, attention_mask, labels)
            loss.backward()
            optimizer.step()
            scheduler.step()
            total_loss += loss.item()

        val_loss, acc, prec, rec, f1, auc = evaluate_model(model, val_loader)
        print(f"Epoch {epoch+1}: ValLoss={val_loss:.4f}, Acc={acc:.4f}, Prec={prec:.4f}, Rec={rec:.4f}, F1={f1:.4f}, AUC={auc:.4f}")

        if val_loss < best_metrics['val_loss']:
            best_metrics.update({'val_loss': val_loss, 'acc': acc, 'prec': prec, 'rec': rec, 'f1': f1, 'auc': auc})
            patience_counter = 0
            best_model_state = model.state_dict()
        else:
            patience_counter += 1
            if patience_counter >= PATIENCE:
                print(f"🛑 Early stopping at epoch {epoch+1}")
                break

    # model_save_path = f"/autodl-tmp/bert_{tag}_{fold_id}_best.pt"
    # torch.save(best_model_state, model_save_path)
    # print(f"✅ Best model ({tag} {fold_id}) saved to {model_save_path}")
    return best_metrics


# ============================================================
# 6. 10-Fold 
# ============================================================
kf = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)
cv_results = []

for fold, (train_idx, val_idx) in enumerate(kf.split(texts, labels)):
    print(f"\n========== CV Fold {fold+1} / 10 ==========")
    metrics = train_one_fold(train_idx, val_idx, texts, labels, fold_id=fold+1, tag="CV")
    cv_results.append(metrics)

cv_df = pd.DataFrame(cv_results)
cv_df.to_csv(OUT_CV, index=False, float_format="%.4f")

print("\n✅ 10-Fold Results saved to:", OUT_CV)
print(cv_df.describe())

# ============================================================
# 7. Leave-One-Project-Out (LOPO)
# ============================================================
project_list = sorted(df['project_name'].unique())
lopo_results = []

for proj in project_list:
    val_idx = df[df['project_name'] == proj].index.tolist()
    train_idx = df[df['project_name'] != proj].index.tolist()

    print(f"\n========== LOPO: Leave [{proj}] Out ({len(val_idx)} samples) ==========")
    metrics = train_one_fold(train_idx, val_idx, texts, labels, fold_id=proj, tag="LOPO")
    metrics['Project'] = proj
    lopo_results.append(metrics)

lopo_df = pd.DataFrame(lopo_results)
lopo_df.to_csv(OUT_LOPO, index=False, float_format="%.4f")

print("\n✅ LOPO Results saved to:", OUT_LOPO)
print(lopo_df.describe())
```
